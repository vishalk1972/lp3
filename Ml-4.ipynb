```python
x_s = 2
step_size = 1
iterations = 0
max_iterations = 100000
rate = 0.01
f = lambda x: (x+3)**2
precision = 0.00001
gd = []

def df(x):
    return 2*(x+3)
```

```python
while precision < step_size and iterations < max_iterations:
    prev = x_s
    iterations+=1
    x_s-=rate*df(prev)
    step_size=abs(x_s-prev)
    gd.append(x_s)
```
